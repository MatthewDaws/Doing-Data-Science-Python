## Bin the data

(Python for Data Analysis, pages 200 and 269).  We create a `cut` object to bin the `Age` column of our DataFrame.  This can then be passed to the `groupby` method to create a GroupBy object.  We show the summary statistics of this.

```python
cats = pd.cut(data["Age"], [-np.Inf,0,18,24,34,44,54,64,np.Inf])
```

```python
grouped = data.groupby(cats)
grouped.describe()
```

We can also more directly copy the sample R script, as follows.

```python
def SiteRange(group):
    return {"length": len(group), "min": group.min(), "mean": group.mean(), "max": group.max()}

data.Age.groupby(cats).apply(SiteRange).unstack()
```

```python
data.groupby(cats).apply(lambda gp: gp.mean()).add_prefix("mean_")
# Or can use
#data.groupby(cats).mean().add_prefix("mean_")
```

From the above, we see that if the user is not "signed in" then we don't have Gender or Age information (which makes sense...)

   - I notice that the Gender mean varies: a notable bias towards teenage men and retired women on the site!  Odd...
   - Very little variation in Impressions count
   - A lot of variation in Click rate

# Some vague thoughts

Above, we looked at the DataFrame `imp_count` which looks at the impression rate against age category:

```python
imp_count.head()
```

In the following, we'll look at each Age Category, ignore extreme values of `Impressions`, normalise, and then plot the Density estimation.  Once we do this, we see that every density is identical!

   - I think this is just an artifact that the data is "generated" (a mixture of two Gaussians?)
   - So basically the impression rate is not so interesting.

```python
fig = plt.figure(figsize=(18,10))
x = imp_count[:17]
for index, agecat in enumerate(x.columns):
    ax = fig.add_subplot(3,3,index + 1)
    imps = x[agecat]
    imps = imps / np.sum(imps)
    #imps.plot(kind="density", ax=ax, legend=agecat)
    imps.plot(ax=ax, legend=agecat)
```

However, if we split between Male and Female, then we do see a difference: it is _exactly_ what I spotted above, namely the "Teenage" and "Senior" categories have rather biased gender ratios, but conditioned on Male of Female, the distribution is the same.

```python
fig = plt.figure(figsize=(18,10))

imp_count_gender = data.groupby(["Impressions", "Gender", cats])
imp_count_gender = imp_count_gender.size().unstack()
for index, agecat in enumerate(imp_count_gender.columns):
    if index == 0: continue # No female/male data if not signed in
    imp_gender = imp_count_gender[agecat].unstack()[:16]
    total_sum = sum( np.sum(imp_gender )) # Normalise across both columns
    imp_gender /= total_sum
    ax = fig.add_subplot(3,3,index)
    #imp_gender.plot(kind="kde", ax=ax, title=agecat)
    imp_gender.plot(ax=ax, title=agecat)
```

## From Click Rate

Recall the `rate` data from above.

```python
fig = plt.figure(figsize=(18,10))

for index, agecat in enumerate(rate.index):
    ax = fig.add_subplot(3,3,index+1)
    rate[agecat].plot(ax=ax, kind="density", title=agecat)
```

This looks a bit odd; I wonder if there is a gender component?

```python
rate = data_has_imps[data_has_imps.Clicks > 0]
rate["ClickRate"] = rate.Clicks / rate.Impressions
rate.Gender = rate.Gender.map({0:"Female", 1:"Male"}) # Pretty this up
rate.describe()
```

```python
rate_cats = pd.cut(rate["Age"], [-np.Inf,0,18,24,34,44,54,64,np.Inf])
grouped = rate.groupby(["ClickRate", "Gender", rate_cats])
ClickRateAge = grouped.size().unstack()
agecat = ClickRateAge.columns[1]

fig = plt.figure(figsize=(18,10))

for index, agecat in enumerate(ClickRateAge.columns):
    if index == 0: continue # Not signed in, so no gender data
    ax = fig.add_subplot(3,3,index)
    PlotData = ClickRateAge[agecat].unstack().fillna(0)
    PlotData /= np.sum(PlotData)
    #TotalSum = sum(np.sum(PlotData))
    #PlotData /= TotalSum
    PlotData.plot(ax=ax,title=agecat) #,kind="density")
```

   - So again we see the markedly different gender profile for the 0--18 and 64+ ages ranges.
   - If you plot with Kernel Density Estimate, then, again, these look like mixtures of Gaussians.
   - Nope: if we condition on gender, then the "Click Rate" looks the same.

# Pick something to check against other days

Let's look at the inpressions required for at least one click through, split by age and gender.

```python
data = pd.read_csv(datapath + "nyt8.csv")

clicks = data[data.Clicks > 0]
clicks.Gender = clicks.Gender.map({0:"Female", 1:"Male"})
clicks_cats = pd.cut(clicks["Age"], [-np.Inf,0,18,24,34,44,54,64,np.Inf])
grouped = clicks.groupby(["Impressions", "Gender", clicks_cats])
clicks_split = grouped.size().unstack()
#agecat = clicks_split.columns[1]
clicks_split.fillna(0)

fig = plt.figure(figsize=(18,10))

for index, agecat in enumerate(clicks_split.columns):
    if index == 0: continue # Not signed in, so no gender data
    ax = fig.add_subplot(3,3,index)
    PlotData = clicks_split[agecat].unstack().fillna(0)
    #PlotData /= np.sum(PlotData)
    TotalSum = sum(np.sum(PlotData))
    PlotData /= TotalSum
    PlotData.plot(ax=ax,title=agecat) # kind="density"
```

We find that this pattern is replicated across each day.
